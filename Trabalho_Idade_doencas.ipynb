```python
dataCV = pd.read_csv('../Trabalho/dados/cause_of_deaths.csv')
dataCV.head()
```

```python
def transpose_table(table):
    return table.transpose()
```

```python
data_g = dataCV.groupby(['Year']).sum()

df = pd.DataFrame(data_g)
transposed_df = transpose_table(df)

data_g_soma = transposed_df.sum(axis=1) 

#print(transposed_df)
data_g_soma.sort_values(0)
#com a funÃ§Ã£o Transpose, ele lista as doenÃ§as por ano, assim podemos identificar a curva de aumento ou diminuiÃ§Ã£o de cada doenÃ§a
```

```python
def max_value_and_column_name(table, row_index):
    max_value = table.iloc[row_index].max()
    max_column = table.columns[table.iloc[row_index].argmax()]
    first_column = table.iloc[row_index][table.columns[0]]
    #first_column = table.index[row_index]
    return first_column, max_column, max_value
```

```python
lista=[]
ii = data_g.shape[0]
i = 0
while i < ii:
    a,b,c = max_value_and_column_name(data_g,i)
    lista.append([a, b,c])
    i += 1
lista

#Pega o principal problema por ano
```
