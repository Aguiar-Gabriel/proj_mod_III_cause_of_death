## VisualizaÃ§Ã£o de quantidade de casos com aumento grave

```python
df_aumento_grave=df_br.loc[:,['Year',"Alzheimer's Disease and Other Dementias","Parkinson's Disease","HIV/AIDS","Drug Use Disorders","Alcohol Use Disorders","Neoplasms","Diabetes Mellitus","Chronic Kidney Disease","Chronic Respiratory Diseases","Digestive Diseases"]]
df_aumento_grave
```

```python
df_aumento_grave.plot( 
    x = 'Year', 
    kind = 'barh', 
    stacked = True, 
    title = 'Stacked Bar Graph', 
    mark_right = True,
    figsize=(40,20))

plt.show()
```

### Visualizando a variaÃ§Ã£o de 3 em 3 anos

```python
b = df_aumento_grave.iloc[::3]
u = df_aumento_grave.iloc[-1]

treshold = b.append(u)
```

```python
treshold.plot( 
    x = 'Year', 
    kind = 'barh', 
    stacked = True, 
    title = 'Stacked Bar Graph', 
    mark_right = True,
    figsize=(40, 20))

treshold_total = treshold.sum(axis=1)
treshold_rel = treshold[treshold.columns[1:]].div(treshold_total, 0)*100
  
for n in treshold_rel: 
    for i, (cs, ab, pc) in enumerate(zip(treshold.iloc[:, 1:].cumsum(1)[n],  
                                         treshold[n], treshold_rel[n])): 
        plt.text(cs - ab / 2, i, str(np.round(pc, 1)) + '%',  
                 va = 'center', ha = 'center', rotation=90, fontsize = 20)

plt.show()
```
