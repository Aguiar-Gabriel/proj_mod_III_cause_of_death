```python
import dash
from dash import dcc
from dash import html
import pandas as pd 
import plotly.express as px    
from dash import Dash, html, dcc, Input, Output
from jupyter_dash import JupyterDash
import jupyter_dash
import plotly.express as px
import plotly.graph_objects as go
from dash.dependencies import Input,Output
```

```python
df = pd.read_csv('cause_of_deaths.csv')
df_as_list = df.values.tolist()
```

```python
df
```

```python
def extrair_dados(df, pais):
  linhas = []

  for coluna in df_as_list:
    if coluna[1] == pais:
      linha = [coluna[2], coluna[12], coluna[17], coluna[9]]
      linhas.append(linha)

  return pd.DataFrame(linhas, columns=["ano", "droga", "alcool", "violencia"])
      

extrair_dados(df_as_list, "CAN")

```

```python
bra = extrair_dados(df_as_list, "BRA")
afg = extrair_dados(df_as_list, "AFG")
can = extrair_dados(df_as_list, "CAN")
```

```python

anos = []
drogas = []
alcool = []
violencia = []
for linha in can.values.tolist():
  anos.append(linha[0])
  drogas.append(linha[1])
  alcool.append(linha[2])
  violencia.append(linha[3])

fig1 = go.Figure()
fig1.add_trace(go.Bar(x=anos, y=drogas, name="Morte por uso de droga", marker=dict(color='black')))
fig1.add_trace(go.Bar(x=anos, y=alcool, name="Morte por uso de alcool", marker=dict(color='RED')))
fig1.add_trace(go.Bar(x=anos, y=violencia, name="Morte por violencia ", marker=dict(color='green')))
```

```python
def grafico_linha_alcool(df ,pais_graf,droga):
  linhas_graf = []

  for coluna in df_as_list:
    if coluna [1] == pais_graf:
      linha_graf = [coluna[1],coluna[2],coluna[12]]
      linhas_graf.append(linha_graf)

  return pd.DataFrame(linhas_graf, columns=["pais_code","pais_graf_ano",droga])

grafico_linha_alcool(df_as_list,"BRA","Drug Use Disorders")
```

```python
def grafico_linha_alcool(df ,pais_graf1,pais_graf2,pais_graf3,droga):
  linhas_graf = []

  for coluna in df_as_list:
    if coluna [1] == pais_graf1 or coluna [1] == pais_graf2 or coluna [1] == pais_graf3:
      linha_graf = [coluna[1],coluna[2],coluna[17]]
      linhas_graf.append(linha_graf)

  return pd.DataFrame(linhas_graf, columns=["codigo","pais_ano",droga])

grafico_linha_alcool(df_as_list,"BRA","AFG","CAN","Causa escolhida")
```

```python

drogasafg = []
drogasbra = []
drogascan = []

for linha in afg.values.tolist():
  drogasafg.append(linha[1])

for linha in bra.values.tolist():
  drogasbra.append(linha[1])

for linha in can.values.tolist():
  drogascan.append(linha[1])


fig = go.Figure()
fig.add_trace(go.Scatter(x=anos, y=drogasbra, name='BRASIL',
                         line=dict(color='green', width=4)))
fig.add_trace(go.Scatter(x=anos, y=drogascan, name='CANADA',
                         line=dict(color='firebrick', width=4)))
fig.add_trace(go.Scatter(x=anos, y=drogasafg, name='AFEGANISTÃO',
                         line=dict(color='black', width=4)))
```

```python
restodrogas = []
restoviolencia = []
restoalcool = []

for coluna in df_as_list:
  if coluna[1] != "BRA" and coluna[1] != "CAN" and coluna[1] != "AFG":
    restodrogas.append(coluna[12])
    restoviolencia.append(coluna[9])
    restoalcool.append(coluna[17])
```

```python
drogassomaafg = 0
drogassomabra = 0
drogassomacan = 0
drogasoutrospaises = 0

for i in drogasafg:
  drogassomaafg += i
for i in drogasbra:
  drogassomabra += i
for i in drogascan:
  drogassomacan += i
for i in restodrogas:
  drogasoutrospaises += i

print(drogassomaafg)
print(drogassomabra)
print(drogassomacan)
print(drogasoutrospaises)
```
